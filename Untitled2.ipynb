## A-1 FACTORIAL

```python
from xmlrpc.server import SimpleXMLRPCRequestHandler, SimpleXMLRPCServer
class Factorial:
    def calculate_factorial(self,num):
        if num<0:
            raise ValueError("Negative")
        result = 0
        for i in range(1,num+1):
            result*=i
        return result
class RequestHandler(SimpleXMLRPCRequestHandler):
    rpc_path='/RPC2'
with SimpleXMLRPCServer(('localhost',8000), requestHandler=RequestHandler) as server:
    server.register_introspection_functions()
    server.register_instance(Factorial())
    print("Ready")
    server.serve_forever()
```

```python
import xmlrpc.client
with xmlrpc.client.ServerProxy("http://localhost:8000/RPC2")as proxy:
    try:
        input_value=int(input())
        result = proxy.calculate_factorial(input_value)
        print(result)
    except Exception as e:
        print(e)
```

## A-3 WORD COUNT

```python
#!/usr/bin/env python3
import sys
for line in sys.stdin:
    linechar = [c for c in line.lower() if c.isalpha() or c==' ']
    words = ''.join(linechar).strip().split()
    for word in words:
        print('%s\t%s' % (word,1))
```

```python
#!/usr/bin/env python3
import sys
from operator import itemgetter
current_word = None
current_count=0
word = None
for line in sys.stdin:
    line=line.strip()
    word, count = line.split('\t',1)
    try:
        count = int(count)
    except ValueError:
        continue
    if current_word==word:
        current_count+=count
    else:
        if current_word:
            print('%s\t%s'%(current_word,current_count))
        current_count=count
        current_word=word
if current_word == word:
    print('%s\t%s'%(current_word,current_count))
```

## A-4 FUZZY

```python
import numpy as np
def union(A,B):
    return np.maximum(A,B)
def intersection(A,B):
    return np.minimum(A,B)
def complement(A):
    return 1-A
def difference(A,B):
    return np.maximum(A,1-B)
A=np.array([0.2,0.4,0.6,0.8])
B=np.array([0.3,0.5,0.7,0.9])
union_result = union(A,B)
intersection_result = intersection(A,B)
complement_result = complement(A)
difference_result = difference(A,B)
print(union_result)
print(intersection_result)
print(complement_result)
print(difference_result)

def cartesian(A,B):
    cartesian=[]
    for i in A:
        for j in B:
            cartesian.append((i,j))
    return cartesian
cartesian_result = cartesian(A,B)
print(cartesian_result)

def min_max(a,b):
    if a.shape[1]==b.shape[0]:
        final = np.empty((a.shape[0],b.shape[1]))
        for i in range(a.shape[0]):
            for j in range(b.shape[1]):
                min_values = np.minimum(a[i, :],b[:, j])
                final[i,j] = np.max(min_values)
        return final
a = np.array([[0.6, 0.5], [0.1, 1], [0, 0.7]])
b = np.array([[0.7, 0.3, 0.4], [0.9, 0.1, 0.6]])
print(min_max(a,b))
```

## A-6 COLONAL SELECTION

```python
import random
pop_size = 100
generations=5
clone_size=10
ab_length=10
mutation_rate = 0.1
population_initialize = lambda size, length: [[random.randint(0,1) for _ in range(length)] for _ in range(size)]
select_best = lambda pop, size: sorted(pop, key=sum, reverse=True)[:size]
mutation = lambda best, rate: [[1-gene if random.random()<rate else gene for gene in antibody] for antibody in best]
def colonal_selection(pop_size,generations, clone_size, ab_length, mutation_rate):
    population = population_initialize(pop_size, ab_length)
    for gen in range(generations):
        best_antibody = select_best(population, clone_size)
        mutated_antibodies = mutation(best_antibody, mutation_rate)
        population+=mutated_antibodies
        print("Generation",gen+1)
        for i, antibody in enumerate(population):
            print(i+1,antibody, sum(antibody))
    return max(population, key=sum), sum(max(population,key=sum))
result, result_affinity = colonal_selection(pop_size, generations, clone_size, ab_length, mutation_rate)
print(result, result_affinity)
```

## B-1 AIRS

```python
import numpy as np
def get_dummy_data(samples=100,features=10):
    data = np.random.rand(samples, features)
    labels = np.random.randint(0,2,size=samples)
    return data, labels
data, labels = get_dummy_data()
split_ratio = 0.8
split_index = int(split_ratio*len(data))
train_data, test_data = data[:split_index],data[split_index:]
train_labels, test_labels = labels[:split_index],labels[split_index:]
def AIRS_train(X,Y):
    detectors = X[np.random.choice(len(X),10,replace=False)]
    return detectors
detectors=AIRS_train(train_data,train_labels)

def AIRS_predict(detectors,X):
    predictions=[]
    for sample in X:
        distances = np.linalg.norm(detectors-sample,axis=1)
        prediction = int(np.argmin(distances))
        predictions.append(prediction)
    return predictions
predictions = AIRS_predict(detectors, test_data)
accuracy = np.mean(predictions==test_labels)
print(accuracy)
```

## B-2 DEAP

```python
import random
from deap import base, creator, tools, algorithms
# Define the evaluation function
def eval_func(individual):
    return sum(x ** 2 for x in individual),
# DEAP setup
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
creator.create("Individual", list, fitness=creator.FitnessMin)
toolbox = base.Toolbox()
# Define attributes and individuals
toolbox.register("attr_float", random.uniform, -5.0, 5.0)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_float, n=3)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)
# Evaluation function and genetic operators
toolbox.register("evaluate", eval_func)
toolbox.register("mate", tools.cxBlend, alpha=0.5)
toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=1, indpb=0.2)
toolbox.register("select", tools.selTournament, tournsize=3)
# Create and evolve the population
population = toolbox.population(n=50)
algorithms.eaSimple(population, toolbox, cxpb=0.5, mutpb=0.1, ngen=20)
# Get the best individual
best_ind = tools.selBest(population, k=1)[0]
best_fitness = best_ind.fitness.values[0]
print("Best individual:", best_ind)
print("Best fitness:", best_fitness)
```

## B-3 ANT COLONY TSP

```python
import numpy as np
import random
distance_matrix = np.array([
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
])
# Parameters for Ant Colony Optimization
num_ants = 10
num_iterations = 50
evaporation_rate = 0.5
pheromone_constant = 1.0
heuristic_constant = 1.0
# Initialize pheromone matrix and visibility matrix
num_cities = len(distance_matrix)
pheromone = np.ones((num_cities, num_cities))  # Pheromone matrix
visibility = 1 / distance_matrix  # Visibility matrix (inverse of distance)
# ACO algorithm
for iteration in range(num_iterations):
    ant_routes = []
    for ant in range(num_ants):
        current_city = random.randint(0, num_cities - 1)
        visited_cities = [current_city]
        route = [current_city]
        while len(visited_cities) < num_cities:
            probabilities = []
            for city in range(num_cities):
                if city not in visited_cities:
                    pheromone_value = pheromone[current_city][city]
                    visibility_value = visibility[current_city][city]
                    probability = (pheromone_value ** pheromone_constant) * (visibility_value ** heuristic_constant)
                    probabilities.append((city, probability))
            probabilities = sorted(probabilities, key=lambda x: x[1], reverse=True)
            selected_city = probabilities[0][0]
            route.append(selected_city)
            visited_cities.append(selected_city)
            current_city = selected_city    
        ant_routes.append(route)
    # Update pheromone levels
    delta_pheromone = np.zeros((num_cities, num_cities))
    for ant, route in enumerate(ant_routes):
        for i in range(len(route) - 1):
            city_a = route[i]
            city_b = route[i + 1]
            delta_pheromone[city_a][city_b] += 1 / distance_matrix[city_a][city_b]
            delta_pheromone[city_b][city_a] += 1 / distance_matrix[city_a][city_b]
    pheromone = (1 - evaporation_rate) * pheromone + delta_pheromone
# Find the best route
best_route_index = np.argmax([sum(distance_matrix[cities[i]][cities[(i + 1) % num_cities]] for i in range(num_cities)) for cities in ant_routes])
best_route = ant_routes[best_route_index]
shortest_distance = sum(distance_matrix[best_route[i]][best_route[(i + 1) % num_cities]] for i in range(num_cities))
print("Best route:", best_route)
print("Shortest distance:", shortest_distance)
```
